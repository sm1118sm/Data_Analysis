# Q4. Pivot table 생성 및 시각화

```python
# Q4. Pivot table 생성 및 시각화
# [사용 데이터] thelook_ecommerce_order.csv'

# 데이터 확인용 코드셀입니다. 참고용으로 활용하시면 됩니다.
df_order = pd.read_csv('thelook_ecommerce_order.csv')
display(df_order.head())
display(df_order.info())
```

<img src="./참고이미지/Q4/Q4_img1.png" width="60%">

- created_at 컬럼을 날짜 형식으로 변경합니다.
- created_at 컬럼을 월 단위로 변환하여 time 컬럼 생성합니다. (dt.to_period('M') 사용).
- time(연-월)과 gender(성별)를 기준으로 user_id의 고유 개수를 집계한 pivot table 생성합니다.
- 고유한 유저 숫자를 세는 것이 목표입니다.

- [결과물] 각 연-월별로 'Female Users'(여성 사용자 수)와 'Male Users'(남성 사용자 수)로 구성된 데이터프레임.

```python
# 해당 셀에 정답 코드를 작성해주세요

import warnings
warnings.filterwarnings('ignore')

df_order['created_at'] = pd.to_datetime(df_order['created_at'])
df_order['time'] = df_order['created_at'].dt.to_period('M')

pivot = pd.pivot_table(
    df_order,
    index = 'time',
    columns='gender',
    values='user_id',
    aggfunc='nunique'
).rename(columns={
    'F': 'Female Users',
    'M': 'Male Users'
})

pivot.columns.name = None
display(pivot)
```

<img src="./참고이미지/Q4/Q4_img2.png" width="60%">

**[시각화]**
- 반환된 pivot table을 막대 그래프로 시각화하여, 시간 흐름에 따른 성별 사용자의 추이를 분석합니다.
- X축: 시간 (연-월 단위), Y축: 사용자 수, 색상: 성별(Female, Male).
- 이 시각화를 통해 시간에 따른 성별 사용자 분포의 변화를 확인할 수 있습니다.

- 구현해야 할 요소
- x축 라벨 회전

```python
# 해당 셀에 정답 코드를 작성해주세요

import matplotlib.pyplot as plt

pivot.plot(kind='bar')
plt.xlabel('time')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# Q6. SQL - 배송 기간별 카테고리 나누기

**[문제조건]**
- 데이터 전처리 작업은 반드시 SQL을 통해 진행해주세요

**[목표]**
- 배송기간별로 데이터를 분류하는 것

**[데이터베이스]**
- SQL & Python 연결 수업 "classicmodels"

**[테이블]**
- ORDERS, ORDERDETAILS

**[수행동작]**
- SALES 컬럼 추가 : priceEach X QUANTITYORDERED를 통해 계산
- SHIPPING_SPEED 컬럼 추가
    - SHIPPEDDATE - ORDERDATE <= 2 "Express"
    - SHIPPEDDATE - ORDERDATE 3일 이상 5일 이하 : "Standard"
    - 그 외 : "Delayed"
- NULL 값에 대한 별도 작업 없음

**[결과물]**
- 작업 결과를 데이터프레임 형태로 출력

```python
# 정답 코드를 작성해주세요

import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
    host='127.0.0.1',
    user='root',
    password='337127ab^^',
    database='classicmodels'
)

join_df = pd.read_sql("""
    SELECT 
        o.orderNumber AS ORDERNUMBER,
        o.orderDate AS ORDERDATE,
        o.shippedDate AS SHPPEDATE,
        DATEDIFF(o.shippedDate, o.orderDate) AS SHIPPING_DAYS,
        CASE
            WHEN DATEDIFF(o.shippedDate, o.orderDate) <= 2 THEN 'Express'
            WHEN DATEDIFF(o.shippedDate, o.orderDate) BETWEEN 3 AND 5 THEN 'Standard'
            ELSE 'Delayed'
        END AS SHPPING_SPEED,
        od.productCode AS PRODUCTCODE,    
        od.priceEach * od.quantityOrdered AS SALES           
    FROM orders o
    JOIN orderdetails od
        ON o.orderNumber = od.orderNumber
""", conn)


display(join_df)

conn.close()
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px

st.set_page_config(layout="wide")
st.title("Q10. Streamlit - 데이터 파일 분석")

orders = pd.read_csv("List of Orders.csv")
details = pd.read_csv("Order Details.csv")

st.subheader("1. List of Orders 파일")
st.dataframe(orders.head(10))

st.subheader("2. List of Orders 파일 분석")

col1, col2 = st.columns(2)

with col1:
    state_btn = st.button("State별 주문 숫자")

with col2:
    city_btn = st.button("City별 주문 숫자")

if state_btn:
    state_cnt = (
        orders.groupby("State")["Order ID"]
        .nunique()
        .reset_index(name="Order Count")
        .sort_values("Order Count", ascending=False)
        .reset_index(drop=True)
    )

    fig = px.bar(
        state_cnt,
        x="State",
        y="Order Count",
        title="Orders by State",
        category_orders={
            "State": sorted(state_cnt["State"])
        }
    )

    st.plotly_chart(fig, use_container_width=True)
    st.dataframe(state_cnt)

if city_btn:
    city_cnt = (
        orders.groupby("City")["Order ID"]
        .nunique()
        .reset_index(name="Order Count")
        .sort_values("Order Count", ascending=False)
        .reset_index(drop=True)
    )

    fig = px.bar(
        city_cnt,
        x="City",
        y="Order Count",
        title="Orders by City",
        category_orders={
            "City": sorted(city_cnt["City"])
        }
    )

    st.plotly_chart(fig, use_container_width=True)
    st.dataframe(city_cnt)

st.subheader("3. Order Details 파일")
st.dataframe(details)

st.subheader("4. Profit 분석")

tab1, tab2 = st.tabs(["Category 기준", "Sub-Category 기준"])

with tab1:
    cat_profit = (
        details.groupby("Category")["Profit"]
        .mean()
        .reset_index()
        .sort_values("Category", ascending=True)
        .reset_index(drop=True)
    )
    st.write("## Category 기준 평균 Profit 집계")
    st.dataframe(cat_profit)

with tab2:
    st.markdown("## Sub-Category 기준 평균 Profit 집계")

    subcat_profit = (
        details.groupby("Sub-Category")["Profit"]
        .mean()
        .reset_index()
        .sort_values("Sub-Category")
        .reset_index(drop=True)
    )

    col_left, col_right = st.columns(2)

    with col_left:
        st.markdown("##### 전체 데이터")
        st.dataframe(subcat_profit)

    with col_right:
        st.markdown("##### 상위 5개 데이터")

        top5 = (
            subcat_profit
            .sort_values("Profit", ascending=False)
            .head(5)
            .reset_index(drop=True)
        )

        st.dataframe(top5)
```
